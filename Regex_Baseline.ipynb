```python
import re
import pandas as pd

# Function to extract age
def extract_age(text):
    age_pattern = re.compile(r'(\b\d{1,3})\s*(?:years?|-year|y/o|years?[- ]old|months?|-month|months?[- ]old)\b', re.IGNORECASE)
    match = age_pattern.search(text)
    
    if match:
        age_value = match.group(1)
        age_unit = "years" if "year" in match.group(0).lower() else "months"
        return age_value + " " + age_unit
    else:
        return 'Unknown'
         
# Function to extract allergies
def extract_allergies(text):
    match = re.search(r'allerg(?:y|ies):(.+?)(?:\.|\,|$)', text, re.IGNORECASE)
    return match.group(1).strip() if match else 'None'

# Function to extract past addictions
def extract_addictions(text):
    matches = re.findall(r'(smok(?:ing|ed)?|alcohol|drug use|addict(?:ion)?)(?:[^.]*?)(?:\.|\,|$)', text, re.IGNORECASE)
    return ', '.join(set(match for match in matches)) if matches else 'None'

# Function to extract gender
def extract_gender(text):
    if re.search(r'\bshe\b|\bher\b|\bhers\b|\bfemale\b', text, re.IGNORECASE):
        return 'Female'
    elif re.search(r'\bhe\b|\bhim\b|\bhis\b|\bmale\b', text, re.IGNORECASE):
        return 'Male'
    else:
        return 'Unknown'

# Function to extract medical conditions
def extract_medical_conditions(text):
    # This can be highly variable and complex depending on the consistency of the records
    # Here's a simple example that looks for common phrases
    conditions = re.findall(r'has\s+(.+?)(?:\,|\sand\s|\.)', text, re.IGNORECASE)
    return ', '.join(set(condition.strip() for condition in conditions)) if conditions else 'None'

# Function to extract drug usage
def extract_drug_usage(text):
    # This will look for phrases commonly associated with medication use
    drugs = re.findall(r'(?:medication[s]?:|take[s]?|taking|prescribed)\s+(.+?)(?:\.|\,|$)', text, re.IGNORECASE)
    return ', '.join(set(drug.strip() for drug in drugs)) if drugs else 'None'

# Assume df is your DataFrame and 'transcription' is the column with the text data

df['Age'] = df['transcription'].apply(extract_age)
df['Allergies'] = df['transcription'].apply(extract_allergies)
df['Past Addictions'] = df['transcription'].apply(extract_addictions)
df['Gender'] = df['transcription'].apply(extract_gender)
df['Medical Conditions'] = df['transcription'].apply(extract_medical_conditions)
df['Drug Usage'] = df['transcription'].apply(extract_drug_usage)

# Display the updated DataFrame
df.head()
```
